```python
df_pop_pivot = pd.read_excel('Task_Data.xlsx', sheet_name = 'Population', skiprows = 1) #МОНГОЛ УЛСЫН ХҮН АМЫН ТОО, бүс, аймаг/нийслэл, хот/хөдөөгөөр
df_lab_pivot = pd.read_excel('Task_Data.xlsx', sheet_name = 'Labor', skiprows = 1) #АЖИЛЛАХ ХҮЧ, хүйс, насны бүлэг, бүс, аймаг, нийслэлээр, улирал, жилээр
df_avgsal_pivot = pd.read_excel('Task_Data.xlsx', sheet_name = 'Avgsal', skiprows = 1) #АЖ АХУЙН НЭГЖ, БАЙГУУЛЛАГЫН АЖИЛЛАГЧДЫН САРЫН ДУНДАЖ ЦАЛИН, бүс, аймаг, нийслэл, хүйс, улирал, жилээр
df_hous_pivot = pd.read_excel('Task_Data.xlsx', sheet_name = 'Housing', skiprows = 1) #БАЙШИН, ГЭР БОЛОН БУСАД СУУЦАНД АМЬДАРДАГ ӨРХ, бүс, аймаг, нийслэл, сум, дүүргээр 
```

```python
df_pop_pivot.rename(columns={'Байршлаар':'Location',
             'Аймаг':'Local','Аймгийн код':'Key'},
             inplace=True)
```

```python
data_pop = df_pop_pivot.melt(id_vars=['Location','Local','Key'], var_name='year', value_name='value')
```

```python
data_pop.replace('^\s+', '', regex=True, inplace=True)
```

```python
data_pop = data_pop[(data_pop['year']==2020) & (data_pop['Location']=='Бүгд') & (data_pop['Key']>5)]
```

```python
df_lab_pivot.rename(columns={'Статистик үзүүлэлт':'Location',
             'Аймаг':'Local','Аймгийн код':'Key','Хүйс':'Sex'},
             inplace=True)
```

```python
df_lab = df_lab_pivot.melt(id_vars=['Location','Local','Key','Sex'], var_name='year', value_name='value')
```

```python
df_lab.replace('^\s+', '', regex=True, inplace=True)
```

```python
data_lab = df_lab[(df_lab['year']=='2021-II') & (df_lab['Location']=='Бүгд') & (df_lab['Key']>4) & (df_lab['Sex']=='Бүгд') & (df_lab['Key']<500)]
```

```python
df_avgsal_pivot.rename(columns={'Аймаг':'Local','Аймгийн код':'Key','Хүйс':'Sex'},
             inplace=True)
```

```python
df_avgsal = df_avgsal_pivot.melt(id_vars=['Local','Key','Sex'], var_name='year', value_name='value')
```

```python
df_avgsal.replace('^\s+', '', regex=True, inplace=True)
```

```python
data_avgsal = df_avgsal[(df_avgsal['year']=='2021-II')]
```

```python
df_hous_pivot.rename(columns={'Засаг захиргааны нэгж':'Local','Статистик үзүүлэлт':'type','2020':'value'},
             inplace=True)
```

```python
df_hous_pivot.replace('^\s+', '', regex=True, inplace=True)
```

```python
data_hous = df_hous_pivot.pivot_table(index="Local", columns="type", values=2020, aggfunc=np.sum)
```

```python
data_hous.reset_index(inplace=True)
```

```python
data_hous['Local'].replace('^\s+', '', regex=True, inplace=True)
```

```python
#data_hous = df_hous_pivot.pivot_table(index="Local", columns="type", aggfunc=np.sum)
```

```python
data_hous
```

```python
df_inner1 = pd.merge(data_pop, data_hous, on='Local')
```

```python
df_inner2 = pd.merge(data_lab, data_avgsal, on='Local')
```

```python
df_inner3 = pd.merge(df_inner1,df_inner2, on = 'Local')
```

```python
df_inner3
```

```python
df_result = df_inner3.groupby(['Local']).agg({'value_x': ['sum'],'value_y': ['sum'],'value': ['sum'],'Байшин': ['sum'],'Бусад': ['sum'],'Гэр': ['sum']})
```

```python
df_result.reset_index(inplace=True)
```

```python
df_result.columns = [col[0] for col in df_result.columns.values]
```

```python
df_result.rename(columns={'value':'Хүн ам (2020)','value_y':'Дундаж цалин (мян.төг, 2021-II улирал)','value_x':'Ажиллагсдын тоо (2021-II улирал)'},
             inplace=True)
```
